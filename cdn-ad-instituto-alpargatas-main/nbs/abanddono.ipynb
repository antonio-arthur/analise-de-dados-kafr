```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar a base
df = pd.read_excel(
    r'C:\Users\anton\Desktop\data_analise_dados\resultado_final_PB_com_abandono.xlsx',
    sheet_name="Sheet1"
)


# Padronizar para maiúsculas
df["rede"] = df["rede"].str.upper()

# Escolher alguns municípios-chave
municipios = ["ALAGOA NOVA", "BANANEIRAS", "CAMPINA GRANDE", "SANTA RITA", "INGÁ"]

# Selecionar colunas de IDEB e abandono
colunas_ideb = [c for c in df.columns if "ideb_" in c]
colunas_abandono = [c for c in df.columns if "taxa_abandono_" in c]

# Filtrar apenas MUNICIPAL
df_filtrado = df[(df["ds_mun_ia"].isin(municipios)) & (df["rede"] == "MUNICIPAL")]

# Garantir uma linha por município
df_filtrado = df_filtrado.drop_duplicates("ds_mun_ia")

# ----- Gráfico do IDEB -----
plt.figure(figsize=(10,6))
anos_ideb = [int(c.split("_")[1]) for c in colunas_ideb]

for mun in municipios:
    subset = df_filtrado[df_filtrado["ds_mun_ia"] == mun]
    if subset.empty:
        print(f"Atenção: {mun} não tem dados para rede MUNICIPAL")
        continue
    valores = subset[colunas_ideb].iloc[0]
    plt.plot(anos_ideb, valores, marker='o', label=mun)

# Linha vertical em 2020
plt.axvline(x=2020, color='red', linestyle='--', linewidth=1.5, label="Atuação Alpargatas (2020)")

plt.title("Evolução do IDEB - Rede Municipal (2005-2023)")
plt.xlabel("Ano")
plt.ylabel("IDEB")
plt.legend()
plt.grid(True)
plt.show()

# ----- Gráfico da Taxa de Abandono -----
plt.figure(figsize=(10,6))
anos_abandono = [int(c.split("_")[2]) for c in colunas_abandono]

for mun in municipios:
    subset = df_filtrado[df_filtrado["ds_mun_ia"] == mun]
    if subset.empty:
        continue
    valores = subset[colunas_abandono].iloc[0]
    plt.plot(anos_abandono, valores, marker='s', label=mun)

# Linha vertical em 2020
plt.axvline(x=2020, color='red', linestyle='--', linewidth=1.5, label="Atuação Alpargatas (2020)")

plt.title("Evolução da Taxa de Abandono Escolar - Rede Municipal (2005-2022)")
plt.xlabel("Ano")
plt.ylabel("Taxa de Abandono (%)")
plt.legend()
plt.grid(True)
plt.show()
```
