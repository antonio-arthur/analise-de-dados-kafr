```python
# Criar uma cópia do df_filtrada para evitar o warning
df_pb = df[df['UF'] == 'PB'].copy()

# Remover espaços do nome da coluna
df_pb.rename(columns=lambda x: x.strip(), inplace=True)

# Converter a coluna POPULAÇÃO ESTIMADA para inteiro
df_pb['POPULAÇÃO ESTIMADA'] = df_pb['POPULAÇÃO ESTIMADA'].str.replace('.', '', regex=False)  # se houver ponto como separador de milhar
df_pb['POPULAÇÃO ESTIMADA'] = df_pb['POPULAÇÃO ESTIMADA'].astype(int)

# Filtrar população entre 20.000 e 22.000
df_pop_filtrada = df_pb[(df_pb['POPULAÇÃO ESTIMADA'] >= 20000) & 
                        (df_pb['POPULAÇÃO ESTIMADA'] <= 22000)]

# Mostrar resultado
df_pop_filtrada[['NOME DO MUNICÍPIO', 'POPULAÇÃO ESTIMADA']]
```

```python
import pandas as pd

# Criar quartis
df_pb['quartil_pop'] = pd.qcut(df_pb['POPULAÇÃO ESTIMADA'], 4, labels=['Q1','Q2','Q3','Q4'])

# Mostrar resultado
df_pb[['NOME DO MUNICÍPIO', 'POPULAÇÃO ESTIMADA', 'quartil_pop']]
```

```python
import pandas as pd

caminho = r"C:\Users\anton\Downloads\POP2020_20220905(Municípios).csv"

# Testar leitura com encoding latino
df = pd.read_csv(caminho, encoding="latin1", sep=";", skiprows= 1)



df = df.rename(columns={"NOME DO MUNICÍPIO": "ds_mun"})

df_filtrada = df[df['UF'] == 'PB']

# Lista dos municípios desejados
municipios_pb = [
    "Alagoa Nova",
    "Bananeiras",
    "Baía da Traição",
    "Cabaceiras",
    "Campina Grande",
    "Caturité",
    "Guarabira",
    "Ingá",
    "Itatuba",
    "João Pessoa",
    "Lagoa Seca",
    "Mogeiro",
    "Queimadas",
    "Santa Rita",
    "Serra Redonda"
]

# Filtrar DataFrame
df_filtradoo = df_filtrada [df['ds_mun'].isin(municipios_pb)]

(df_filtradoo)
```

```python
tabela_ia = pd.read_excel(r'C:\Users\anton\Desktop\data_analise_dados\resultado_final_PB.xlsx')

tabela_ia
```

```python
tabela_final = pd.merge(df_filtradoo, tabela_ia, on="ds_mun", how="left")

tabela_final

```

```python
# Salvar tabela_final em Excel
tabela_final.to_excel(r'C:\Users\anton\Desktop\tabela_final.xlsx', index=False)

print("Arquivo salvo com sucesso!")
```
